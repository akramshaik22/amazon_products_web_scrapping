```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv

driver = webdriver.Chrome()
driver.minimize_window()

'''
Given Url = https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1
By using above url i get the base url = https://www.amazon.in/s?k=bags 
By using this base url we can increment page no's to get the next page url.
'''

base_url = 'https://www.amazon.in/s?k=bags&page='

f = open("product_info.csv", 'w')
writer = csv.writer(f)
writer.writerow(['URL', 'NAME', 'PRICE', 'RATING', 'NO_OF_REVIEWS', 'ASIN', 'MANUFACTURER', 'PRODUCT_DESCRIPTION'])

url_hits = 0

for i in range(1,21):
    print("\n\n ##### PAGE -- ", i, end="\n\n")
    for retry in range(3):
        url = base_url + str(i)
        driver.get(url)
        product_url = driver.find_elements(By.CSS_SELECTOR,"div.a-section div.a-section h2 a")
        names = driver.find_elements(By.CSS_SELECTOR,"div.a-section div.a-section h2 a span")
        rating = driver.find_elements(By.CSS_SELECTOR,"div.a-section div.a-section div span[aria-label] span.puis-normal-weight-text")
        reviews = driver.find_elements(By.CSS_SELECTOR,"div.a-section div.a-section div span[aria-label] a span.a-size-base")
        review_url = driver.find_elements(By.CSS_SELECTOR,"div.a-section div.a-section div span[aria-label] a")
        prices = driver.find_elements(By.CSS_SELECTOR,"div.a-section div a span.a-price span span.a-price-whole")
        price_url = driver.find_elements(By.CSS_SELECTOR,"div.a-section div a")

        names_cnt =  len(names)
        prices_cnt = len(prices)
        reviews_cnt = len(reviews)
        rating_cnt = len(rating)
        print(names_cnt, prices_cnt, rating_cnt, reviews_cnt)

        url_dict = {}
        
        if 0 not in [names_cnt, prices_cnt, rating_cnt, reviews_cnt]:

            ## PART-1
            '''
            All Products URLs and details in 20 pages are stored in url_dict dictionary by having their url as key.
            Part-1 of Assignment is Completed.
            '''

            for j in range(names_cnt):
                url_dict[product_url[j].get_attribute("href")] = [names[j].text]

            for j in range(prices_cnt):
                try:
                    url_dict[price_url[j].get_attribute("href")].append(prices[j].text)
                except:
                    pass

            for j in range(len(review_url)):
                try:
                    url_dict[review_url[j].get_attribute("href")[:-16]] += [rating[j].text, reviews[j].text]
                except:
                    pass


            ## PART-2
            '''
            Here By hitting every url in url_dict we collect the ASIN, MANUFACTURER, PRODUCT DESCRIPTION.
            After collecting all information the data will be stored in 'product_info.csv' file as below format:

            URL    NAME   PRICE   RATING   NO_OF_REVIEWS   ASIN   MANUFACTURER   PRODUCT_DESCRIPTION

            '''


            for i in url_dict:
                url_hits += 1
                asin, manufacturer, description = ' ', ' ', ' '
                driver.get(i)
                try:
                    product_details = driver.find_elements(By.CSS_SELECTOR,"div#detailBullets_feature_div ul li span span")
                    # print(len(product_details))
                    manufacturer = product_details[5].text
                    asin = product_details[7].text
                except:
                    pass
                try:
                    description = driver.find_elements(By.CSS_SELECTOR,"div#productDescription p span")[0].text
                except:
                    pass

                # print(manufacturer, asin)
                print(i, url_dict[i], asin, manufacturer)
                
                try:
                    
                    if len(url_dict[i]) < 2:
                        writer.writerow([i] + url_dict[i] + [' ', ' '] + [asin, manufacturer, description])
                    else:
                        writer.writerow([i] + url_dict[i] + [asin, manufacturer, description])
                except:
                    print("!!! CANT WRITE ROW")
                    
        else:
            print("!!!!! RETRY -----", retry+1)
            continue
        
        break
print("*** TOTAL NO OF URL HITS = ", url_hits)
f.close()
driver.quit()
    
    

    
```
